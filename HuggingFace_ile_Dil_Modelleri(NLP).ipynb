```python
from transformers import pipeline

# Model ve tokenizer yükleme
generator = pipeline('text-generation', model='gpt2')
```

```python
# Prompt
prompt = "I think this training is useful for you"

# Çıktıyı üretme
outputs = generator(prompt, max_length=50, truncation=True, pad_token_id=50256)  # 50256 = GPT-2'nin EOS token ID'si
print(outputs[0]["generated_text"])
```

```python
generator=pipeline("text-generation", model="distilgpt2")
```

```python
# Başlangıç metni
prompt = "Explain the benefits of artificial intelligence in healthcare:"
```

```python
# Çıktı üretme
outputs = generator(
    prompt,
    max_length=100,  # Daha uzun çıktı için artırılabilir
    truncation=True,
    pad_token_id=50256,  # Uyarıyı engeller
    repetition_penalty=1.2  # Döngü ve tekrarları azaltır
)

# Üretilen metni yazdır
print(outputs[0]["generated_text"])
```

Metin Özetleme

```python
summary=pipeline("summarization")
```

```python
# Text to summarize
text = """
Artificial intelligence (AI) has become an integral part of modern technology, revolutionizing industries such as healthcare, finance, education, and entertainment. AI refers to the simulation of human intelligence processes by machines, particularly computer systems. These processes include learning, reasoning, problem-solving, perception, and language understanding. In recent years, AI has made significant strides, with advancements in machine learning and natural language processing. These technologies enable computers to analyze large amounts of data, make predictions, and automate tasks, which have greatly improved efficiency and decision-making. However, with the rapid development of AI, there are also concerns about job displacement, privacy issues, and ethical dilemmas related to AI decision-making. As AI continues to evolve, it is important to strike a balance between innovation and responsibility.
"""
```

```python
# Model Output
outputs = summary(text, max_length=100, min_length=30, do_sample=False)
```

```python
print(outputs[0]["summary_text"])
```

Metin Benzerliği Karşılaştırması

```python
from sentence_transformers import SentenceTransformer, util

# Modeli yükle
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Karşılaştırmak istediğiniz iki metin
text1 = "I have always enjoyed exploring new places, especially when it comes to visiting cities with rich histories. Recently, I traveled to Paris, where I had the chance to explore iconic landmarks such as the Eiffel Tower, the Louvre Museum, and the beautiful streets of Montmartre. The experience was unforgettable, and I can't wait to plan my next adventure."
text2 = "Exploring new cities and places is something I find incredibly exciting. A few weeks ago, I visited Paris and had the opportunity to visit famous landmarks such as the Eiffel Tower, the Louvre, and Montmartre. It was a memorable experience, and I am already thinking about my next trip."

# Metinleri vektörlere dönüştür
embeddings1 = model.encode(text1, convert_to_tensor=True)
embeddings2 = model.encode(text2, convert_to_tensor=True)

# Benzerliği hesapla
similarity = util.pytorch_cos_sim(embeddings1, embeddings2)

# Sonuç
print(f"Benzerlik Skoru: {similarity.item()}")
```
